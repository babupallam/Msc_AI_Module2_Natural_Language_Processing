### Automatic Tagging

The Default Tagger. The simplest possible tagger assigns the same tag to each token (NN). Notice poor performance.

```python
import nltk
from nltk.corpus import brown

# Define a raw text string
raw = 'I do not like green eggs and ham, I do not like them Sam I am!'

# Tokenize the raw text using nltk's word_tokenize function
tokens = nltk.word_tokenize(raw)

# Create a default tagger that tags every word as a noun ('NN')
default_tagger = nltk.DefaultTagger('NN')

# Tag the tokens using the default tagger
#  It applies the default_tagger to the tokens list. This means it assigns the tag 'NN' (noun) to each word in the tokens list.
default_tagger.tag(tokens)

# Get tagged sentences from the Brown corpus (news category)
brown_tagged_sents = brown.tagged_sents(categories='news')

# Evaluate the accuracy of the default tagger on the Brown corpus and print the result
print('\n Perc of Correctly tagged tokens -> ', round(default_tagger.evaluate(brown_tagged_sents)*100),'%')
```

Explanation of the output:
- The output Perc of Correctly tagged tokens -> 13 % means that the default_tagger, which simply tags every word as a noun ('NN'), achieved an accuracy of 13% when evaluated against the Brown corpus.
- 13% accuracy: This low accuracy indicates that simply tagging every word as a noun is not a very effective strategy. In real-world text, only about 13% of words are actually nouns.

Note:

- The tagged words evaluated against each category in brown corpus results different accuracy.
- the reason is that
  - Different genres have different word distributions: Each category in the Brown corpus represents a different genre of text (news, fiction, scientific, etc.). These genres have distinct writing styles, vocabulary, and grammatical structures. For example, news articles may use more formal language and have a higher frequency of nouns and verbs, while fiction may contain more adjectives and adverbs.
  
  - Part-of-speech tag frequencies vary across genres: As a result of the different word distributions, the frequency of different parts of speech (nouns, verbs, adjectives, etc.) will vary across categories. A tagger trained on one genre may not perform as well on another if the distribution of tags is significantly different.

Creating and evaluating lookup tagger with a range of sizes.
(Read 4.3   The Lookup Tagger in https://www.nltk.org/book/ch05.html)

---



```python
#from the book sextion 4.3:

# Calculate the frequency distribution of words in the "news" category of the Brown corpus.
fd = nltk.FreqDist(brown.words(categories='news')) # this wont be list type

# Calculate the conditional frequency distribution of tags for each word in the "news" category.
# This distribution will store the frequency of each tag for each word.
cfd = nltk.ConditionalFreqDist(brown.tagged_words(categories='news'))

print(cfd)

# Get the 100 most frequent words from the frequency distribution.
most_freq_words = fd.most_common(100)

# Create a dictionary mapping words to their most likely tag.
# Iterate over the most frequent words and for each word, get the tag with the highest frequency from the conditional frequency distribution.
likely_tags = dict((word, cfd[word].max()) for (word, _) in most_freq_words)

# Create a unigram tagger using the dictionary of most likely tags as the model.
# This tagger will assign tags to words based on the most likely tag for each word.
baseline_tagger = nltk.UnigramTagger(model=likely_tags)

# Evaluate the performance of the baseline tagger on a set of tagged sentences.
# The brown_tagged_sents variable should contain a list of sentences where each sentence is a list of (word, tag) tuples.
# The evaluate method will return the accuracy of the tagger on the given sentences.
baseline_tagger.evaluate(brown_tagged_sents)
```

```python

# Define a function to evaluate the performance of a baseline tagger
def performance(cfd, wordlist):
    # Create a dictionary mapping words to their most frequent tag
    lt = dict((word, cfd[word].max()) for word in wordlist) # keys are words from wordlist and values are the most frequent tag for that word, obtained from a conditional frequency distribution cfd.

    # Create a unigram tagger using the dictionary and a default tagger
    baseline_tagger = nltk.UnigramTagger(model=lt, backoff=nltk.DefaultTagger('NN'))

    # Evaluate the tagger on the "news" category of the brown corpus and return the accuracy
    return baseline_tagger.evaluate(brown.tagged_sents(categories='news'))
```

```python

# Define a function to display the performance of a lookup tagger with varying model sizes
def display():
    # Import the pylab library for plotting
    import pylab
    # Create a frequency distribution of words from the "news" category of the brown corpus
    words_by_freq = list(nltk.FreqDist(brown.words(categories='news')))
    # Create a conditional frequency distribution of tagged words from the "news" category
    cfd = nltk.ConditionalFreqDist(brown.tagged_words(categories='news'))
    # Define the model sizes to test
    sizes = 2 ** pylab.arange(15)
    # Calculate the performance for each model size
    perfs = [performance(cfd, words_by_freq[:size]) for size in sizes]
    # Plot the performance against the model size
    pylab.plot(sizes, perfs, '-bo')
    # Set the plot title
    pylab.title('Lookup Tagger Performance with Varying Model Size')
    # Display the plot
    pylab.show()

display()
```

### N-Gram Tagging

Unigram Tagging. For each token, assign the tag that is most likely for that particular token. Training of a UNIGRAM tagger and evaluation


```python
from nltk.corpus import brown
brown_tagged_sents = brown.tagged_sents(categories='news')
brown_sents = brown.sents(categories='news')
unigram_tagger = nltk.UnigramTagger(brown_tagged_sents)
unigram_tagger.tag(brown_sents[2007])
print('\n Perc of Correctly tagged tokens -> ', round(unigram_tagger.evaluate(brown_tagged_sents)*100),'%')
```

Separating training and testing data sets. Notice the different scores between train (used to build system 90%) and test (not used while building the system 10%).

```python
size = int(len(brown_tagged_sents) * 0.9)
size
train_sents = brown_tagged_sents[:size]
test_sents = brown_tagged_sents[size:]
unigram_tagger = nltk.UnigramTagger(train_sents)
print('\n')
print('\n Perc of Correctly tagged tokens on training data -> ', round(unigram_tagger.evaluate(train_sents)*100),'%')
print('\n Perc of Correctly tagged tokens on testing data  -> ', round(unigram_tagger.evaluate(test_sents)*100),'%')
```

Bigram tagger. Notice perfomance difference between training and testing sentence.

```python
bigram_tagger = nltk.BigramTagger(train_sents)
bigram_tagger.tag(brown_sents[2007])
unseen_sent = brown_sents[4203]
bigram_tagger.tag(unseen_sent)


print('\n')
print('\n Perc of Correctly tagged tokens on training data -> ', round(bigram_tagger.evaluate(train_sents)*100),'%')
print('\n Perc of Correctly tagged tokens on testing data  -> ', round(bigram_tagger.evaluate(test_sents)*100),'%')
```
