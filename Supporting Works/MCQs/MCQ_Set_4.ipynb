### **Multilayer Perceptron (MLP) Basics**

1. **What does an MLP primarily consist of?**  
   a) Convolutional layers  
   b) Input, hidden, and output layers  
   c) Recurrent layers  
   d) Residual connections  
   **Answer:** b  
   *Explanation:* MLP consists of an input layer, one or more hidden layers, and an output layer.

2. **What function does a single neuron in an MLP apply to its input?**  
   a) Multiplication  
   b) Weighted sum and activation function  
   c) Subtraction  
   d) Division  
   **Answer:** b  
   *Explanation:* A neuron computes a weighted sum of the inputs and applies an activation function to produce an output.

   b) Introduces non-linearity to the model  
   c) Sets weights to zero  
   d) Normalizes inputs

   *Explanation:* The activation function introduces non-linearity to the model, allowing it to solve complex problems.

   b) To offset the weighted sum  
   c) To decrease the output  
   d) To add noise to the input  
   **Answer:** b  
   *Explanation:* The bias term helps adjust the output by offsetting the weighted sum, improving the model's flexibility.

5. **Which activation function is commonly used in hidden layers of MLPs?**

   d) Sigmoid

    *Explanation:* A larger stride reduces the output size, as the kernel moves further with each step, covering fewer regions of the input.

    *Explanation:* Padding is used to maintain the same output size as the input by adding extra values (usually zeros) around the edges of the input.
